##  1. 随机数
python中产生随机数的主要有两个模块`random`和`np.random`

### 1.1 指定形状的随机数
- `numpy.random.rand(d0, d1, ..., dn)`  
Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1).


````python
import numpy as np

print("shape with one dimension")
print(np.random.rand(1))

print("shape with two dimension")
print(np.random.rand(2, 3))

print("shape with three dimension")
x = np.random.rand(2, 3, 2)
print(x)
print(x.shape)
````

### 1.2 指定范围内指定大小的随机数
- `numpy.random.randint(low, high=None, size=None, dtype='l')`  
Return random integers from low (inclusive) to high (exclusive).
Return random integers from the “discrete uniform” distribution of the specified dtype in the “half-open” interval [low, high). If high is None (the default), then results are from [0, low).

````python
# 在范围[4, 10)之间产生20个随机数字
print(np.random.randint(low=4, high=10, size=20))

# 在范围[4, 10)之间产生个三维大小随机数字
print(np.random.randint(low=4, high=10, size=(2, 3, 2)))
````

### 1.3 从指定的范围内取出相应的数据
- `numpy.random.choice(a, size=None, replace=True, p=None)`
Generates a random sample from a given 1-D array

````python
# 从[1, 3, 5]中产生一个随机数字
print(np.random.choice([1, 3, 5], size=1))

# 从[1, 3, 5, 7, 9, 11]中产生一个二维矩阵数字（有放回）
print(np.random.choice([1, 3, 5, 7, 9, 11], size=(2, 3), replace=True))

# 从[1, 3, 5, 7]中产生一个二维矩阵数字（有放回）,每个元素取得的概率[0.05, 0.05, 0.1, 0.8]
print(np.random.choice([1, 3, 5, 7], size=(4, 5), replace=True, p=[0.05, 0.05, 0.1, 0.8]))
````

### 2.1 中文&ax.plot&xlabel rotation
总结点    

- 安装中文包   
安装包`pyplotz`，使用时   

```
pltz = PyplotZ()
pltz.enable_chinese()

pltz...
```

- 使用ax.plot和指定figsize  

```
fig, ax = plt.subplots(figsize=(17, 8))
```

- 更改xlabel和ylabel的字体大小和倾斜度   
ax.get_xticklabels


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyplotz.pyplotz import PyplotZ
from io import StringIO

sns.set()
pltz = PyplotZ()
pltz.enable_chinese()


s = """
|day|valid_set_xgb|valid_set_bef|
| 20190403 | 0.934962| 0.914719| 
| 20190402 | 0.942149  | 0.922432|
| 20190401 | 0.942458| 0.923098| 
| 20190331 | 0.943411| 0.927580| 
|20190330 | 0.938604| 0.919794| 
|20190329 | 0.94628| 0.925569| 
|20190328 | 0.938888| 0.914644| 
|20190327 | 0.940326| 0.921115| 
| 20190326 | 0.942518 | 0.920653 | 
| 20190325 | 0.942286 | 0.920422 | 
| 20190324 | 0.937038 | 0.915979|
| 20190323 | 0.935666 | 0.912570|
| 20190322 | 0.933774 |  0.911209|
| 20190321 | 0.937109 |  0.918083|
| 20190320 | 0.938866 |  0.918264|
| 20190319 | 0.938786 |  0.916669| 
| 20190318 | 0.934995 |  0.913706|
| 20190317 | 0.931763 | 0.910795 |
| 20190316 | 0.928204 | 0.908476 |
"""
dd = pd.read_csv(StringIO(s), sep="|", dtype={"day": str})
dd = dd[dd.columns[1:-1]]


fig, ax = plt.subplots(figsize=(17, 8))

ax.plot(dd.day[::-1], dd.valid_set_xgb[::-1], "r-o", label="xgb model")
ax.axhline(dd.valid_set_xgb.mean(), linestyle="--", color="red", label="xgb model mean")

ax.plot(dd.day[::-1], dd.valid_set_bef[::-1], "b-o", label="curr model")
ax.axhline(dd.valid_set_bef.mean(), linestyle="--", color="blue", label="curr model mean")

ax.set_xlabel("day", fontsize=16)
ax.set_ylabel("AUC", fontsize=16)
# ax.set_title("xgb gjxc comparision", fontsize=20)
pltz.title("中文标题")
pltz.title_size(20)
# ax.set(xlabel="day", ylabel="valid AUC", title="xgb gjxc comparision")
ax.legend(fontsize="large")
ax.grid()

# Set the tick labels font
for label in ax.get_xticklabels():
    label.set_fontname('Arial')
    label.set_fontsize(16)
    label.set_rotation(30)
for label in ax.get_yticklabels():
    label.set_fontname("Arial")
    label.set_fontsize(16)
    
fig.show()
````

### 2.3 另外方式渲染中文

除了上面介绍到使用包`pltz`，可以下载相应的中文字体，然后绘图指定字体即可。发现在使用`plzt`包时，发现其功能和原生的`axes`有区别，仍存在一定问题，故而需求另外渲染中文的方式。

- 方式1    
1.字体安装   
```
$ sudo apt-get install fonts-wqy-zenhei # 安装字体, 
$ cd /usr/share/fonts/truetype/wqy/ # cd到字体安装目录下
$ fc-list :lang=zh  # 查看新安装的字体的路径，在下面fname中添加
```
2.指定字体      
```
import matplotlib.pyplot as plt
import matplotlib.font_manager as mfm
zh_font = mfm.FontProperties(fname="/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc")
plt.xlabel(u'测试一下 ', fontsize=20, fontproperties=zhfont) # specify font 
```

- 方式2 
```
plt.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体设置
plt.rcParams['axes.unicode_minus'] = False
```

> Note: 方式1在linux系统中测试通过，方式2测试失败；两种方式均未在windows系统中测试
